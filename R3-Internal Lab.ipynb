# Q8 Split the pima_df into training and test set in the ratio of 70:30 (Training:Test).</b>


```python
X = df.drop('class', axis=1)
y = df ['class']
x_train, x_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=0) #70/30 split
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

# Q9 Create the decision tree model using â€œentropyâ€ method of reducing the entropy and fit it to training data.</b>



```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(criterion = 'entropy', max_depth = 2)
```

```python
dt_model.fit(x_train, y_train)
```

# Q10 Print the accuracy of the model & print the confusion matrix 

```python
dt_model.score(x_test, y_test)
```

Print the feature importance of the decision model - Optional

```python
y_pred = dt_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test, y_pred))
```
